Import Packages

```python
import pandas as pd
import numpy as np
import warnings

from datetime import datetime, timedelta
from IPython.display import display

warnings.filterwarnings("ignore")
```

Read Data with Pandas

```python
pop_data_raw = pd.read_csv("Pop_Data_Raw.csv")#read input data

pop_transposed = pop_data_raw.transpose()#flip dataframe

final_table = pop_transposed.iloc[2:3,0:12]#create data frame header columns

count_start = 0
count_end = 12
for i in range(9):
    transposed_country = pop_transposed.iloc[4:, count_start: count_end]
    transposed_country.columns = range(transposed_country.columns.size)
    final_table = pd.concat([final_table, transposed_country], axis=0, ignore_index=True)
    count_start = count_end 
    count_end = count_end + 12

final_table.head(1)
```

Create Population Dataframe

```python

final_table.columns=["Population_15To64_Percent", "Population_65andUp_Percent","Population_Growth_Percent", "Population_Female_Percent", "Population_Male_Percent", "Rural_Population_Percent", "Rural_Population_Growth_Percent", "Urban_Population_Percent", "Urban_Population_Growth_Percent", "Life_Exp_Birth_Total_Yr", "Life_Exp_Birth_Male_Yr", "Life_Exp_Birth_Female_Yr"]
```

Add Country Name and Year

```python

count = 5
iterator = 0;
countries = ["Canada", "USA", "Mexico", "Burundi", "Chad", "Niger", "Bolivia", "Philippines", "Indonesia"]

country_year = pd.DataFrame(columns=["Country","Year"])

pop_data = {}
pop_data["Country"] = "Blank"
pop_data["Year"] = 0
  
country_year = country_year.append(pd.DataFrame([pop_data]), ignore_index=True)


for i in range(144):
    if count>20:
        count = 5
        iterator = iterator + 1

    pop_data["Year"] = 2000+count
    pop_data["Country"] = countries[iterator]
    count = count + 1
    country_year = country_year.append(pd.DataFrame([pop_data]), ignore_index=True)

display(country_year.tail(5))
```

Join Country_Year with Final_Table

```python
final_table = pd.concat([country_year, final_table], axis=1)
display(final_table.tail(20))
```

Fill in Empty Cells

```python

#Replace emply values with NaN
final_table['Life_Exp_Birth_Female_Yr'].replace({'..': np.NaN}, inplace = True)
final_table['Life_Exp_Birth_Male_Yr'].replace({'..': np.NaN}, inplace = True)
final_table['Life_Exp_Birth_Total_Yr'].replace({'..': np.NaN}, inplace = True)

#Replace NaN values with last observation
final_table.fillna(method="ffill", inplace=True)


display(final_table.head(20))
```

Remove Original Labels

```python
pop_data = final_table.drop(0)
pop_data.head(200)
```

Generate CSV

```python
final_table.to_csv("Population_Processed_Table.csv", index = False, header = True)
```
